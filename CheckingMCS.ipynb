**Function created to block each MCS of the model considered using the COBRApy *bounds* method and find an optimal solution to the problem using the function *slim_optimize()***

```python
def check(case):
    with model:
        for r in case:
            model.reactions.get_by_id(r).bounds=[0,0]
        sol=model.slim_optimize(error_value=-1)
        if not sol==-1:
            print(case,"Fallo")
```
